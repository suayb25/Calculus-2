Volumes found by rotating a region around a line are called solids of revolution.

```python
#@title
%pylab inline
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
fig, ax = plt.subplots()
ax.axis("off")
img = mpimg.imread('volumes.png')
imgplot = plt.imshow(img)
plt.show()
```

# **When rotating around x-axis or any horizontal line**

For solids of revolution, the cross sections have the shape of a **disk** or the
shape of a **washer**.

A = $\pi r^2$,    

A = $\pi r_{outer}^2 - \pi r_{inner}^2$

The area of the cross-sections can be described with the formulas

$\int_{a}^{b} A(x) \,dx$

Disk =  $\int_{a}^{b} \pi r^2 \,dx$

Washer = $\int_{a}^{b} \pi r_{outer}^2 - \pi r_{inner}^2 \,dx$

# **When rotating around y-axis or any vertical line**

$\int_{c}^{d} A(y) \,dy$

Disk =  $\int_{c}^{d} \pi r^2 \,dy$

Washer = $\int_{c}^{d} \pi r_{outer}^2 - \pi r_{inner}^2 \,dy$

# **Example** 

Consider the region bounded by the curve $y =\sqrt[3]{x}$, the x-axis, and the line x = 8. What is the volume of the solid of revolution formed by rotating this region around the x-axis?


```python
#@title
fig, ax = plt.subplots()
ax.axis("off")
img = mpimg.imread('volumes2.png')
imgplot = plt.imshow(img)
plt.show()
```

Cross-sections -> Disks

V = $\int_{0}^{8} \pi r^2 \,dx$

r = y = $y =\sqrt[3]{x}$

V = $\int_{0}^{8} \pi (\sqrt[3]{x})^2 \,dx$


= $\pi \int_{0}^{8}  {x}^{2/3} \,dx$


= $\pi \frac{3}{5} {x}^{5/3} |_0^8$

= $\pi \frac{3}{5} {8}^{5/3} - 0$
