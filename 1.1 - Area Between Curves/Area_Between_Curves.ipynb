**Recall: To compute the area below a curve y = f(x), between x = a and x = b, we can divide up the region into rectangles.**

```python
#@title
%pylab inline
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
fig, ax = plt.subplots()
ax.axis("off")
img = mpimg.imread('area_under_curve.png')
imgplot = plt.imshow(img)
plt.show()
```

**Width of a rectangle:** Δx

**Sample point on x-axis for rectangle i:** ${x_i}^*$

**Height of rectangle:** $f({x_i}^*)$

**The area of one small rectangle is**
base.height

Δx.$f({x_1}^*)$, Δx.$f({x_2}^*)$, ... ,Δx.$f({x_i}^*)$, ... ,Δx.$f({x_n}^*)$

**n =** # of rectangles

**The approximate area under the curve is**


$\sum_{i=1}^{n} Δx.f({x_i}^*)$

**Area =** $\lim_{n=\infty}{\sum_{i=1}^{\infty}f({x_i}^*).Δx}$ = $\int_{a}^{b} f(x) \,dx$

------------------------------

To compute the area between the curves y = f(x) and y = g(x), between x = a and
x = b, we can divide up the region into rectangles.


```python
#@title
fig, ax = plt.subplots()
ax.axis("off")
img = mpimg.imread('area_under_curve_2.png')
imgplot = plt.imshow(img)
plt.show()
```

**Width of each rectangle:** Δx

**Sample point in interval i:** ${x_i}^*$

The area of one small rectangle is

Δx. $(f({x_i}^*)-g({x_i}^*))$

The approximate area between the two curves is

$\sum_{i=1}^{n} Δx.(f({x_i}^*)-g({x_i}^*))$

The exact area between the two curves is

$\lim_{n=\infty}{\sum_{i=1}^{\infty}(f({x_i}^*)-g({x_i}^*)).Δx}$ **=** **$\int_{a}^{b} f(x)-g(x) \,dx$**

This formula works as long as **f(x) $\geq$ g(x).**

# **Example: Find the area between the curves $y = x^2 + x$ and $y = 3 − x^2$**

```python
#@title
fig, ax = plt.subplots()
ax.axis("off")
img = mpimg.imread('area_under_curve_example.png')
imgplot = plt.imshow(img)
plt.show()
```

```python
import numpy as np
import scipy.integrate as integrate
import scipy.special as special
```

```python
def function1(x):
  return x**2 + x

def function2(x):
  return 3 - x**2

def differences(x):
  return function2(x)-function1(x)
```

$x^2 + x - (3 - x^2) = 2x^2 + x -3$ 

```python
coeff = [2, 1, -3] 
equation_solve = np.roots(coeff)
print(equation_solve)
```

```python
result = integrate.quad(differences, equation_solve[0], equation_solve[1])
```
